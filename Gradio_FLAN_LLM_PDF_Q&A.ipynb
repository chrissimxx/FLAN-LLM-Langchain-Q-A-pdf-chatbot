```python
!pip install gradio
!pip install langchain
!pip install PyPDF2
!pip install huggingface_hub
!pip install unstructured
%pip install unstructured[local-inference]

from langchain import PromptTemplate, HuggingFaceHub, LLMChain
from langchain.document_loaders import UnstructuredURLLoader
from langchain.document_loaders import PDFMinerLoader
from langchain.document_loaders import TextLoader
from PyPDF2 import PdfReader
import gradio as gr

api_key="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" #Hugging face API

def get_llm_response(pdf_text, question):
    template="Question: {question}\n{answer}"
    prompt = PromptTemplate(template=template, input_variables=["question", "answer"])
    llm_chain = LLMChain(prompt=prompt, llm=HuggingFaceHub(repo_id="google/flan-t5-xl", model_kwargs={"temperature":0, "max_length":4000}, huggingfacehub_api_token=api_key))
    response = llm_chain.run(question=question, answer=pdf_text)
    return response

def extract_pdf_text(pdf_file):
    reader = PdfReader(pdf_file) # Read the uploaded PDF file. Upload your file in collab or read from URL
    pdf_text = ""
    print("Total pages=",len(reader.pages)) # Total pages in the pdf
    page_numbers_to_read = [0] # Specify which pages you want to read

    # Read the given pages
    for page in page_numbers_to_read:
        page_text = reader.pages[page].extract_text()
        print("Reading pages {} of :\n{}".format(page+1, str(len(reader.pages))))
        #print("The contents of page {} are:\n{}".format(page+1, page_text))
        pdf_text += page_text

    #print("All extracted text:\n", pdf_text)

    #save the extracted data from pdf to a txt file
    file1=open(r"pdf2text.txt","a")
    file1.writelines(pdf_text)
    print("file saved")
    return pdf_text

# Set up Gradio interface
def pdf_question_answering(pdf_file, question):
    pdf_text = extract_pdf_text(pdf_file.name)
    response = get_llm_response(pdf_text, question)
    return response

iface = gr.Interface(fn=pdf_question_answering, inputs=["file", "text"], outputs="text",
                     title="PDF Question Answering", 
                     description="Extracts text from a PDF file and answers a given question using a language model")

# launch the interface
iface.launch(debug=True)
```
